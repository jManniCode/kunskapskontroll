# Korrelationer

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Läs in data
df = pd.read_csv("diamonds.csv")

# Skriv ut antal rader före rensning
print("Antal rader före rensning:", df.shape[0])

# Ta bort rader med saknade värden
df = df.dropna()
print("Efter borttagning av rader med saknade värden:", df.shape[0])

# Ta bort diamanter med 0 i någon dimension (fysiskt orimligt)
df = df[(df["x"] > 0) & (df["y"] > 0) & (df["z"] > 0)]
print("Efter borttagning av diamanter med 0 i x, y eller z:", df.shape[0])
```

```python

# Välj bara numeriska variabler
plot_vars = ["price", "carat", "x", "y", "z", "depth", "table"]

# Skapa korrelationsmatris
corr_matrix = df[plot_vars].corr()

# Lista starkaste korrelationerna (absolutvärde, exkludera 1.0-diagonalen)
corr_pairs = (
    corr_matrix.where(~corr_matrix.isna() & (corr_matrix != 1))
    .stack()
    .abs()
    .sort_values(ascending=False)
    .reset_index()
)
corr_pairs.columns = ["Var1", "Var2", "Korrelationskoefficient"]

# Ta bort dubbletter (Var1/Var2 == Var2/Var1)
corr_pairs["combo"] = corr_pairs.apply(lambda row: tuple(sorted([row["Var1"], row["Var2"]])), axis=1)
corr_pairs = corr_pairs.drop_duplicates(subset="combo")

# Visa de starkaste sambanden
print(corr_pairs.head(10))
```

# Heatmap

```python
corr = df[["price", "carat", "x", "y", "z", "depth", "table"]].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Heatmap")
plt.show()
```

# Korrelation pris och karat

```python
# Scatterplot mellan karat och pris
plt.figure(figsize=(8, 5))
plt.scatter(df["carat"], df["price"], alpha=0.1)
plt.title("Pris i relation till karat")
plt.xlabel("Karat")
plt.ylabel("Pris (USD)")
plt.show()
```
